# Recopilación de datos
## Es un historia de clientes telecom_churn.csv
* gender
* tenure: Cantidd de meses que el cliente usa el servicio
* montlyCharges: cargos mensuales
* totalCharges: Total pagado por el cliente
* contract: tipo de contrato del cliente
* InternetServices: tipo de servicio de internet contratado
* Churn: abandono

# Importación de librerías

```python
import pandas as pd
import numpy as np

# Visualización
import matplotlib.pyplot as plt

# Preprocesamiento
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# División de datos
from sklearn.model_selection import train_test_split

# Modelo
from sklearn.ensemble import RandomForestClassifier

# Métricas
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# Validación
from sklearn.model_selection import cross_val_score

print("\n LIBRERÍAS CARGADAS CORRECTAMENTE !!!!!!!")
```

# Carga del DataSet

```python
datos = pd.read_csv("telecom_churn.csv")

# Mostrar primeras filas
display(datos.head())
# Mostrar ultimas filas
display(datos.tail())
# Mostrar aleatorio
display(datos.sample(5))
```

# Exploración inicial

```python
print("\n INFORMACION DEL DATASET")
print("===================================\n")
print(datos.info())

print("\n ESTADISTICAS DEL DATA SET")
print("===================================\n")
display(datos.describe())

print("\n VALORES NULOS POR CAMPO")
print("===================================\n")
print(datos.isnull().sum())
```

# 2. Limpieza y preprocesamiento
## En datos reales existen
* valores nulos
* errores
* duplicados
* inconsistentes
* ruido

```python
datos = datos.dropna()
print("\n VALORES NULOS ELIMINADOS ----->\n")
```

## Eliminación de duplicados

```python
datos = datos.drop_duplicates()
print("\n VALORES DUPLICADOS ELIMINADOS ====>\n")
```

## Conversión de variables categóricas

```python
encoder = LabelEncoder()

columnas_categoricas = [
    "gender",
    "Contract",
    "InternetService"
]

for columna in columnas_categoricas:
    datos[columna] = encoder.fit_transform(datos[columna])

display(datos.tail())
print("== REALIZADO LA CONVERSION EN: gender, Contract y InternetService ==")
```

# Selección de variables
## Variables predictoras

```python
X = datos[
    [
        "gender",
        "tenure",
        "MonthlyCharges",
        "TotalCharges",
        "Contract",
        "InternetService"
    ]
]
print("\n Seleccionado...! las variables independientes ====>")
```

## Variable objetivo

```python
y = datos["Churn"]
print("\n Definido---! la variable dependiente ====>")
```

## Normalización de datos

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
print("\n APLICADO LA NORMALIZACION AL DATA SET")
```

## 3. División de datos
### Lo dividimos en
* Entrenamiento
* Prueba
### Generalmente:
* 80% entrenamiento
* 20% prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
print("\n REALIZADO LA DIVISION 80% PARA EL ENTRENAMIENTO, 20% PARA PRUEBA>")
```

## 4. Entrenamiento del algoritmo
### En esta etapa se aprende patrones automáticamente
### Se usará RandomForest
* Alta precisión
* Robustez
* Capacidad para manejar múltiples valores
* Reducción de overfitting
## Creación del modelo

```python
modelo = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
print("\n MODELO CREADO.....!")
```

## Entrenamiento

```python
modelo.fit(X_train, y_train)
print("\n MODELO ENTRENDO.....!\n")
```

Durante esta fase:
* El algoritmo analiza ejemplos
* Ajusta parámetros internos
* Identifica patrones
* Genera árboles de decisión
* Optimiza predicciones

## 5. Generación del modelo
Despues del entrenamiento se obtiene un modelo entrenado capaz de predecir nuevos casos
## Realizar predicciones

```python
predicciones = modelo.predict(X_test)

print("\n PREDICCION.....!, DE LOS PRIMEROS.....")
print(predicciones[:7])
```

Si sale 1: posible abandono; con 0: cliente estable

## Ejemplo de predicción individual

```python
# LOS DATOS CORRESPONDEN A (gender,tenure,MonthlyCharges,TotalCharges,Contract,InternetService)

nuevo_cliente = pd.DataFrame([{
    'gender': 1,
    'tenure': 12,
    'MonthlyCharges': 80,
    'TotalCharges': 1200,
    'Contract': 0,
    'InternetService': 1
}])

nuevo_cliente = scaler.transform(nuevo_cliente)

resultado = modelo.predict(nuevo_cliente)

print("\n POSIBLE ESTADO DEL NUEVO CLIENTE ===> ",resultado)
```
