# Limpieza básica

````python
texto = """
Hola!!!
Este es un texto de prueba.

Visita: https://ejemplo.com
"""

texto = re.sub(r"\n", " ", texto)

texto = re.sub(
    r"[^a-zA-Z0-9áéíóúñü\s]",
    " ",
    texto
)

texto = re.sub(r"\s+", " ", texto).strip()

print(texto)
````

## Normalización

````python
def normalize_text(texto: str) -> str:

    texto = texto.lower()

    texto = re.sub(r"[^\w\s]", " ", texto)

    texto = re.sub(r"\d+", " ", texto)

    texto = re.sub(r"\s+", " ", texto).strip()

    texto = unicodedata.normalize("NFKD", texto)

    texto = "".join(
        c for c in texto
        if not unicodedata.combining(c)
    )

    return texto


texto = "¡Minería de Textos 2025 en la UPEA!"

texto_normalizado = normalize_text(texto)

print(texto_normalizado)
````

# Eliminación de ruido

````python
texto = """
Síguenos en https://ejemplo.com
#MineriaDeTextos @usuario
"""

# Eliminar URLs
texto = re.sub(r"https?:\/\/\S+", "", texto)

# Eliminar hashtags (ej: #MineriaDeTextos)
texto = re.sub(r"#\w+", "", texto)

# Eliminar menciones (ej: @usuario)
texto = re.sub(r"@\w+", "", texto)

# Limpiar espacios extras
texto = re.sub(r"\s+", " ", texto).strip()

print(texto)
````

# Remoción de stopwords

````python
nltk.download("punkt_tab")
nltk.download("stopwords")
````

````python
texto = "Este es un ejemplo de texto para eliminar stopwords."

tokens = word_tokenize(texto.lower())

stopwords_spanish = set(
    stopwords.words("spanish")
)

tokens_sin_stop = [
    token
    for token in tokens
    if token not in stopwords_spanish
]

print(tokens_sin_stop)
````

# Tokenización

````python
texto = """
La minería de textos es poderosa.
Permite extraer patrones útiles.
"""

tokens_palabras = word_tokenize(texto.lower())

tokens_oraciones = sent_tokenize(texto)

print("Tokens por palabra:")
print(tokens_palabras)

print("\nTokens por oración:")
print(tokens_oraciones)
````

# Lematización

````python
lemmatizer = WordNetLemmatizer()

words = [
    "running",
    "cars",
    "better"
]

for word in words:

    lemma = lemmatizer.lemmatize(word)

    print(f"{word} -> {lemma}")
````

# DTM Document-Term matrix

````python
corpus = [
    "la minería de textos es poderosa",
    "la minería de textos permite extraer patrones"
]

vectorizer = CountVectorizer()

dtm = vectorizer.fit_transform(corpus)

print("Vocabulario:")
print(vectorizer.get_feature_names_out())

print("\nDTM:")
print(dtm.toarray())
````

````python
tdm = dtm.T

print("TDM (término × documento):")
print(tdm.toarray())
````

# Frecuencia de términos

````python
frecuencia_terminos = dtm.toarray().sum(axis=0)

print("Frecuencia de términos:")
print(frecuencia_terminos)
````
